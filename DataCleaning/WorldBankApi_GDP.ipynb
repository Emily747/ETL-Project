```python

# Dependencies
import requests
import json
import os
import pandas as pd
import numpy as np
from pprint import pprint
```

```python
#Getting country codes
countrycodes_file = "Resources/CountryCodesLookup.csv"
countrycodes = pd.read_csv(countrycodes_file, encoding="ISO-8859-1")
countrycodes = countrycodes.rename(columns={'COUNTRY':'country'})
countrycodes = countrycodes.rename(columns={'A2 (ISO)':'countryCode'})
countrycodes["GDP"]=''
countrycodes["POP"]=''
#NewColumn
worldGDP=countrycodes[['country','countryCode','GDP']].copy()
worldGDP.head()
```

```python
import requests
import json

Base_Url="http://api.worldbank.org/v2/country/"
Country = ""

for index, row in worldGDP.iterrows():
    try:
        Country = str.lower(row['countryCode'])

        GDP_url = Base_Url + Country + '/indicator/NY.GDP.PCAP.CD?date=2017&format=json'
        POP_url= Base_Url + Country + '/indicator/SP.POP.TOTL?date=2017&format=json'




        #full syntax
        GDP_response = requests.get(GDP_url)
        GDP_response_json = GDP_response.json()
        POP_response = requests.get(POP_url)
        POP_response_json = POP_response.json()
#         print(Country)
#         print (GDP)
#         print(POP)
#         print(index)

        GDP= GDP_response_json[1][0]['value']
        worldGDP.loc[index,"GDP"]=GDP
        
        
        POP= POP_response_json[1][0]['value']
        worldGDP.loc[index,"POP"]=POP
        
    except: 

        print("Missing field/result... skipping.")

```
